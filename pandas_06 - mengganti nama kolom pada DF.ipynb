```python
import pandas as pd 
import numpy as np
```

```python
# persiapan DF
n_rows = 5
n_cols = 5
cols = tuple('ABCDE')

df = pd.DataFrame(np.random.randint(1, 10, size=(n_rows, n_cols)), columns=cols)
df
```

```python
# mengganti label sebuah kolom dengan fungsi rename
df.rename(columns={'C':'Hobi'})
```

```python
# mengganti label banyak kolom dengan fungsi rename dan dict
df.rename(columns={'A':'Nama', 'B':'Alamat', 'D':'Kota'})
```
