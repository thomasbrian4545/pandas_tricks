```python
import pandas as pd 
import numpy as np 
```

```python
# persiapan DF
n_rows = 5
n_cols = 5
cols = tuple('ABCDE')

df = pd.DataFrame(np.random.randint(1, 10, size=(n_rows, n_cols)), columns = cols)
df
```

```python
df1 = df.copy(deep=True)
df1 = df1.drop([1,2])
df1
```

```python
df2 = df.copy(deep=True)
df2 = df2.drop([3,4])
df2
```

```python
# merger dengan inner
df_inner = pd.merge(df1, df2, how='inner')
df_inner
```

```python
# merger dengan outer
df_outer = pd.merge(df1, df2, how='outer')
df_outer
```
