```python
import pandas as pd
import numpy as np
```

```python
# persiapan DF
n_rows = 10
n_cols = 1
cols = ('usia',)

df = pd.DataFrame(np.random.randint(1, 99, size=(n_rows, n_cols)), columns=cols)
df
```

```python
# pengelompokkan nilai numerik menggunakan cut()
df['kelompok_usia'] = pd.cut(df['usia'],
                             bins=[0, 18, 65, 99],
                             labels=['anak', 'dewasa', 'manula'])
df
```
