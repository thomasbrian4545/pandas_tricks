```python
import pandas as pd 
import numpy as np
```

```python
# persiapan DF
n_rows = 10
n_cols = 5
cols = tuple('ABCDE')

df = pd.DataFrame(np.random.randint(1, 20, size=(n_rows, n_cols)), columns=cols)
df
```

```python
# seleksi kolom dan baris menggunakan loc
df.loc[[0,3,4], ['B','E']]
```

```python
# seleksi baris dengan kondisi
df.loc[df['B'] > 10, ['B','D','E']]
```

```python
# slicing dengan loc
df.loc[0:4, ['C', 'E']]
```
