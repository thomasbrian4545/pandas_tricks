```python
import pandas as pd 
import numpy as np
```

```python
# persiapan DF
n_rows = 10
n_cols = 5
cols = tuple('ABCDE')

df = pd.DataFrame(np.random.randint(1, 20, size=(n_rows, n_cols)), columns=cols)
df
```

```python
# membagi DF menjadi dua secara acak dengan proporsi tertentu
df.shape
```

```python
proporsi = 0.7
df_1 = df.sample(frac=proporsi)
df_2 = df.drop(df_1.index)

print(f'df_1 shape: {df_1.shape}')
print(f'df_2 shape: {df_2.shape}')
```
