```python
import pandas as pd
```

```python
df = pd.read_csv('./data/Iris.csv')
df.head()
```

```python
# mengenal groupby() dan fungsi agregasi
df.groupby('Species')['PetalLengthCm'].mean().to_frame()
```

```python
# agregasi dengan fungsi agg()
df.groupby('Species')['PetalLengthCm'].agg(['mean', 'median', 'count'])
```

```python
# agregasi dengan fungsi describe()
df.groupby('Species')['PetalLengthCm'].describe()
```
