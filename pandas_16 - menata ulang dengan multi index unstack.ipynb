```python
import pandas as pd
```

```python
# persiapan DF
df = pd.read_csv('./data/titanicfull.csv')
df.head()
```

```python
# DF dengan multi index dari hasil groupping
df.groupby(['pclass', 'sex'])['survived'].mean().to_frame()
```

```python
# menata ulang DF dengan multi index
df.groupby(['pclass', 'sex'])['survived'].mean().unstack()
```
